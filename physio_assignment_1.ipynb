# 1. Fibonacci (Progressive Exercise Plan)


A.

Input and Output
Input: An integer n representing the current week of the rehabilitation plan (e.g., n = 5 for Week 5).

Output: An integer representing the target number of exercise repetitions the patient needs to complete that week.


B.
Python Implementation

```python
# 1. Recursive Approach
def fibonacci_recursive(n):
    # Base cases based on the assignment's progression example
    if n == 1:
        return 1
    elif n == 2:
        return 2

    # Recursive call
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

# 2. Dynamic Programming Approach (Tabulation)
def fibonacci_dp(n):
    # Base cases
    if n == 1:
        return 1
    elif n == 2:
        return 2

    # Create an array to store the repetitions for each week
    # We use size (n + 1) so the array index matches the week number
    reps = [0] * (n + 1)
    reps[1] = 1
    reps[2] = 2

    # Iterate to build the sequence from the bottom up
    for i in range(3, n + 1):
        reps[i] = reps[i - 1] + reps[i - 2]

    return reps[n]

# Test the algorithms
week = 5
print(f"Recursive: For Week {week}, the patient should do {fibonacci_recursive(week)} reps.")
print(f"DP: For Week {week}, the patient should do {fibonacci_dp(week)} reps.")
```

C.
Time Complexity

Recursive:
Best: $O(1)$ (Only if n is 1 or 2, hitting the base case immediately).

Average / Worst: $O(2^n)$. This is exponential time. It is highly inefficient because it continuously recalculates the exact same weeks over and over again as the tree branches out.

Dynamic Programming:Best / Average / Worst: $O(n)$. This is linear time. It only calculates each week's repetitions exactly once, storing the answer in the reps array to look up later.


D.
Space Usage

Recursive: $O(n)$. While it doesn't create arrays, it uses memory on the "call stack." If you ask for Week 100, it opens 100 nested functions in memory before it can finish.

Dynamic Programming: $O(n)$. We create a list (reps) of size n to store the answers. (Note: This can actually be optimized to $O(1)$ by only storing the last two variables instead of the whole list, but the array method makes the concept easiest to explain!)

E.
Link to Physiotherapy Practice

In physiotherapy, tissue healing and strength building require a concept called "progressive overload." If a patient does the exact same exercises forever, they plateau. The Fibonacci sequence models a safe, non-linear progression plan. In the early weeks (Week 1 to 3), the increase in repetitions is small and gradual, allowing the injured area to adapt. As the weeks go on and the patient regains strength, the volume of repetitions scales up much faster, safely pushing the patient toward full mobility.

# 2. Sorting Algorithms (Patient Pain Levels)

A. Input and Output

Input: An unsorted list of integers representing patient pain scores (e.g., [8, 5, 6, 2, 9, 4]).

Output: The same list sorted in ascending order (e.g., [2, 4, 5, 6, 8, 9]).

B. Python Implementation

```python
# Sample Dataset: Patient pain scores (1-10 scale)
pain_scores = [8, 5, 6, 2, 9, 4]

# 1. Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        swapped = False
        # Last i elements are already in place, so we don't check them
        for j in range(0, n - i - 1):
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        # If no two elements were swapped by inner loop, array is sorted
        if not swapped:
            break
    return arr

# 2. Selection Sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        # Find the minimum element in the remaining unsorted array
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j

        # Swap the found minimum element with the first element of the unsorted part
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# 3. Insertion Sort
def insertion_sort(arr):
    # Traverse from the second element to the end
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        # Move elements of arr[0..i-1], that are greater than key,
        # to one position ahead of their current position
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Test the algorithms (using copies of the array so they don't pre-sort for each other)
print(f"Original Pain Scores: {pain_scores}")
print(f"Bubble Sort Result: {bubble_sort(pain_scores.copy())}")
print(f"Selection Sort Result: {selection_sort(pain_scores.copy())}")
print(f"Insertion Sort Result: {insertion_sort(pain_scores.copy())}")
```

C. Time Complexity

Bubble Sort: * Best: $O(n)$ (Occurs when the array is already sorted, thanks to the swapped flag optimization).

Average / Worst: $O(n^2)$ (Occurs when the array is reverse sorted).

Selection Sort:
Best / Average / Worst: $O(n^2)$. Selection sort is rigid; it always scans the entire remaining array to find the absolute minimum, regardless of whether the array is already sorted.

Insertion Sort:
Best: $O(n)$ (Occurs when the array is already sorted, as the while loop condition fails immediately).

Average / Worst: $O(n^2)$ (Occurs when elements must be continuously shifted, especially in a reverse-sorted array).

D. Space Usage

Best / Average / Worst Space Complexity for all three: $O(1)$.

Explanation: These are all "in-place" sorting algorithms. They do not require creating new arrays or massive data structures in memory to perform the sort. They simply swap variables around within the original list, making them highly memory efficient.

E. Link to Physiotherapy Practice

In a busy physiotherapy clinic, therapists handle dozens of patients daily. If a clinic director pulls a weekly report of patient pain scores or balance test results, the data is naturally unsorted. Applying a sorting algorithm allows the clinic to instantly organize the data to perform triage. For example, by sorting pain scores in descending order, the therapist can immediately identify and prioritize patients experiencing severe pain (scores of 8, 9, or 10) for urgent follow-up consultations or adjusted pain management strategies.

# 3. Binary Search (Patient Record Lookup)

A.
Input and Output

Input: A sorted list of integers representing Patient IDs (e.g., [102, 105, 210, 304, 415, 590]) and a target ID to search for (e.g., 304).

Output: The index position of the target ID within the list (e.g., index 3). If the patient ID is not found in the database, the algorithm returns -1.


B. Python Implementation


```python
# Sample Dataset: A perfectly sorted list of Patient IDs
patient_ids = [102, 105, 210, 304, 415, 590, 602, 711, 890]

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        # Find the middle index
        mid = (low + high) // 2

        # Check if the target is present at mid
        if arr[mid] == target:
            return mid

        # If target is greater, ignore the left half
        elif arr[mid] < target:
            low = mid + 1

        # If target is smaller, ignore the right half
        else:
            high = mid - 1

    # Target was not found in the array
    return -1

# Test the algorithm
target_patient = 415
result_index = binary_search(patient_ids, target_patient)

if result_index != -1:
    print(f"Success: Patient ID {target_patient} found at index {result_index}.")
else:
    print(f"Error: Patient ID {target_patient} not found in the database.")
```

C. Time Complexity

Best: $O(1)$. This happens if the target ID happens to be the exact middle element of the array on the very first check.

Average / Worst: $O(\log n)$. Logarithmic time is incredibly fast. Instead of checking every single patient record one by one (which would be $O(n)$), Binary Search cuts the searchable database in half with every single step.

D. Space Usage

Space Complexity: $O(1)$ (Constant space).

Explanation: Because we implemented this iteratively using a while loop, we only needed to create a few variables (low, high, mid) to act as pointers. We did not duplicate the array or use the call stack, making it highly memory efficient.

E. Link to Physiotherapy Practice

While a small private clinic might only have a few dozen patients, a major hospital's physiotherapy and rehabilitation department will have thousands or tens of thousands of patient records. If a therapist needs to pull up the rehabilitation history for Patient #84920, searching sequentially from the beginning (Linear Search) is a massive waste of computing resources. Because hospital databases assign ascending patient IDs, using Binary Search allows the system to retrieve the patient's exercise history and recovery timeline almost instantaneously, streamlining administrative tasks and letting the therapist focus entirely on patient care.

# **TASK 3**

A. Measuring Execution Time and Memory

B. Generating Data & Plotting Execution Time

```python
import time
from memory_profiler import memory_usage
import numpy as np
import matplotlib.pyplot as plt

# --- Profiling Function ---
def profile_algorithm(algorithm_func, data):
    # 1. Measure Memory (returns memory consumed in MiB)
    mem_usage = memory_usage((algorithm_func, (data.copy(),)), max_usage=True)

    # 2. Measure Time
    start_time = time.perf_counter()
    algorithm_func(data.copy())
    end_time = time.perf_counter()

    execution_time = end_time - start_time
    return execution_time, mem_usage

# --- Data Generation & Profiling Loop ---
# 1. Define the input sizes to test
input_sizes = [100, 500, 1000, 2000, 3000]

# 2. Create lists to store the execution times for the y-axis
bubble_times = []
selection_times = []
insertion_times = []

print("Starting profiling... this may take a few moments.")

# 3. Run the profiling loop
for size in input_sizes:
    # Generate an array of random patient pain scores (1 to 10)
    test_data = np.random.randint(1, 11, size).tolist()

    # Profile Bubble Sort
    b_time, b_mem = profile_algorithm(bubble_sort, test_data)
    bubble_times.append(b_time)

    # Profile Selection Sort
    s_time, s_mem = profile_algorithm(selection_sort, test_data)
    selection_times.append(s_time)

    # Profile Insertion Sort
    i_time, i_mem = profile_algorithm(insertion_sort, test_data)
    insertion_times.append(i_time)

    print(f"Processed size {size}: Bubble ({b_time:.4f}s), Selection ({s_time:.4f}s), Insertion ({i_time:.4f}s)")

# --- Visualization ---
# 4. Plotting the results using matplotlib
plt.figure(figsize=(10, 6))

# Plot each algorithm's execution time
plt.plot(input_sizes, bubble_times, label='Bubble Sort', marker='o')
plt.plot(input_sizes, selection_times, label='Selection Sort', marker='s')
plt.plot(input_sizes, insertion_times, label='Insertion Sort', marker='^')

# Add labels, title, and formatting
plt.title('Execution Time vs Input Size (Sorting Patient Pain Scores)')
plt.xlabel('Input Size (Number of Patient Records)')
plt.ylabel('Execution Time (Seconds)')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

C. Discussing the Trade-offs

Bubble Sort: It is incredibly easy to understand and write, but it is notoriously inefficient for large datasets because of the sheer number of swaps it performs.

Selection Sort: While it also scales poorly, it minimizes the number of actual swaps compared to Bubble Sort (it only swaps once per pass). However, it is rigid and always checks every single element, even if the array is already mostly sorted.

Insertion Sort: For real-world data that might already be partially sorted (like adding a few new patient files to an existing alphabetical database), Insertion Sort performs significantly better than the other two, leaning closer to its best-case $O(n)$ time.
